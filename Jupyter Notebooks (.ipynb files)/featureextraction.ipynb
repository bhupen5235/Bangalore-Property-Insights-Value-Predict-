```python
# Function to  Extract number of bathrooms

import re
def extract_bathrooms(text):
    
    match = re.search(r'(\d+)\s*Bathrooms?', text, re.IGNORECASE)
    if match:
        num = int(match.group(1))  # Get the matched number as an integer 
        if num > 15:
            return int(str(num)[-1])  # If greater than 15, return the last digit
        else:
            return num  # If less than or equal to 15, return the number itself
    else:
        return None  # Return None if no match is found
```

```python
df['Details']=df['Details'].astype(str)
```

```python
df['Bathrooms'] = df['Details'].apply(extract_bathrooms)
df["Bathrooms"].shape
```

Using Google Gemini to extract the Region from locality

```python
#Using Google Gemini to find the major areas of banglore

import google.generativeai as genai
import time
genai.configure(api_key="YOUR API KEY")
model = genai.GenerativeModel('gemini-1.5-flash')   # Version of Gemini
```

```python
def categorize_location(locality):
    # giving Prompt to the Gemini to extract region
    
    response = model.generate_content(f"Identify the geographical region that is '{locality}' of the given Bangalore neighborhood. Limit your response to one of these ten regions: North Bangalore, South Bangalore, East Bangalore, West Bangalore, Central Bangalore, Northeast Bangalore, Southeast Bangalore, Northwest Bangalore, Southwest Bangalore, or Peripheral areas. For example, if I say 'Marathahalli', you should respond with 'East Bangalore'.")
    location = response.text.lower()
    time.sleep(2)  

    # Taking relevent information from the responce given by Google Gemini
    if "south bangalore" in location:    
        return "South Bangalore"
    
    elif "north bangalore" in location:
        return "North Bangalore"
    
    elif "east bangalore" in location:
        return "East Bangalore"
    
    elif "west bangalore" in location:
        return "West Bangalore"
    
    elif "central bangalore" in location:
        return "Central Bangalore"
    else:
        return "Peripheral areas"
  
```

```python
# Spliting list into smaller parts, so that Gemini resource not get exhausted

l1=localities[:25]   # trying with smaller list
l2=localities[25:251]
l3=localities[251:]

print(f'l1--{len(l1)}\n'
      f'l2--{len(l2)}\n'
      f'l3--{len(l3)}')

# Ensuring that all element are present
print("Total number of unique localities--",len(localities))
print(f'total data --{len(l1)+len(l2)+len(l3)}')
```

```python
# Defining dictionaries to store the responce from Google Gemini
area_mapping1={}
area_mapping2={}
area_mapping3={}
```

```python

for locality in l1:
    area = categorize_location(locality)
    area_mapping1[locality] = area
    time.sleep(2)
# Converting responce to dictionary
responce1= pd.DataFrame(list(area_mapping1.items()), columns=['Locality', 'Region'])


for locality in l2:
    area = categorize_location(locality)
    area_mapping2[locality] = area
    time.sleep(2)
# Converting responce to dictionary
responce2=pd.DataFrame(list(area_mapping2.items()), columns=['Locality', 'Region'])


for locality in l3:
    area = categorize_location(locality)
    area_mapping3[locality] = area
    time.sleep(2)
# Converting responce to dictionary
responce3=pd.DataFrame(list(area_mapping3.items()), columns=['Locality', 'Region'])
```

```python
# Saving Responce to CSV file to use later.
# As generating responce for such large number of inputs is long process.

responce1.to_csv("region1.csv")
responce2.to_csv("region2.csv")
responce3.to_csv("region3.csv")
```

Extacting from Property Column

```python
# extracting no of BHK and property type i.e. Villa, Apartment etc.
```

```python
df.head()
```

```python
def extract_bhk_and_type(text):
    words = text.split()  # Split the string into words
    bhk = ' '.join(words[:2])  # Join the first two words (e.g., "4 BHK")
    property_type = words[2]  # The third word is the property type (e.g., "Villa" or "Apartment")
    return bhk, property_type
```

```python
# Creating new columns to store BHK and property type
df[['BHK', 'Property_Type']] = df['Property'].apply(extract_bhk_and_type).apply(pd.Series)
```
